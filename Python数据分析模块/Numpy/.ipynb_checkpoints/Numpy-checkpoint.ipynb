## 二、ndarray创建

### 1、使用**np.array()**由Python的列表进行创建
- ndarray默认所有元素类型一致；
- 若传进的参数元素数据类型不一致，则自动将其转化为同类型数据，优先级:str > float > int

````python
lst = [1, 2, 3, 4, 5]

nd1 = np.array(lst)
nd2 = np.array([[1, 2],
                [3, 4]])
nd3 = np.array(list('ABC'))
nd4 = np.array([1, 2, 'a']) # 根据优先级，自动将其中的数字转化为字符串
 
print(nd1, nd2, nd3, nd4)
````

#### 1、np.ones
- np.ones(shape, dtype= None, order = 'C')，创建元素全为1的数组
 - shape用于指定数组的维度；
 - dtype用于指定数组元素的类型

````python
np.ones((3, 3), dtype = 'float64')
````

````python
ones = np.ones((100, 80, 3), dtype = float) # 元素全为1，显示为白色
plt.imshow(ones)
plt.show()
````

#### 2、np.zeros
- np.zeros(shape, dtype= None, order = 'C')，创建元素全为0的数组
 - shape用于指定数组的维度；
 - dtype用于指定数组元素的类型

````python
np.zeros((4, 4), dtype = int)
````

#### 3、np.full
- np.full(shape, fill_value, dtype=None, order='C')，创建以指定元素填充的数组
 - shape用于指定数组的维度；
 - fill_value用于指定填充的元素；
 - dtype用于指定数组元素的类型

````python
np.full((4, 4), fill_value = 2, dtype = float)
````

#### 4、np.eye
- np.eye(N, M=None, k=0, dtype=<class 'float'>, order='C')，创建对角线为1的数组
 - N用于指定数组的行数；
 - M用于指定数组的列数，不指定时默认为与参数N相同；
 - k表示对角线索引，默认为主对角线索引0，主对角线上方索引为正数，主对角线下方索引为负数，通过设定k值分别在主对角线、上方或下方填充元素1；
 - dtype用于指定数据类型

````python
np.eye(N = 5, M = 4, k = 0)
````

````python
np.eye(N = 5, M = 4, k = 3)  # 主对角线上方索引为正
````

#### 5、np.linspace
- np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None)，创建指定起始值以及长度的数组
 - start用于指定起始值；
 - stop用于指定终止值；
 - num用于指定生成数组的长度，默认长度为50；
 - endpoint用于设定终止值是否包含在数组中，若为True则包含在数组中，反之则不包含在数组中；
 - retstep用于设定是否输出步长，若为True则输出步长，反之则不输出步长；
 - dtype用于指定数组元素的类型；

````python
np.linspace(10, 60, num = 10, endpoint = True, retstep = True)
````

## 四、ndarray的基本操作

### 1、索引
- 一维数组的索引与列表索引完全相同；
- 二维及以上的数组索引则是逐层索引直至获取对应元素；
- 通过索引可实现对数组中元素的修改；

````python
n = np.random.random(size = (2, 3))
n
````

````python
# 第一个[0]表示第一行，第二个[0]表示第一行中第一个元素，以下两种方法等价
n[0][0], n[0, 0]
````

### 2、切片
- 从原始数组中切取指定的数据，切片主要根据索引实现；

````python
# 逗号前边表示维度，逗号后边表示对应维度中的元素
print(n[0, 0:3], n[1, 0:3], end = '\n')
````

````python
# 只切取第一行
n[0:1]
````

````python
n1 = np.random.random(size = (4, 3, 3))
n1
````

````python
# 切取第一维所有数据
n1[0]
````

````python
# 切取第一维和第二维所有数据
n1[0:2]
````

````python
# 切取第一维和第二维中第一行和第二行的所有数据
n1[0:2, 0:2]
````

````python
# 切取第一维和第二维中第一行和第二行的前两个数
n1[0:2, 0:2, :2]
````

### 4、重塑
- reshape(shape)，重新指定数组的维度
 - shape用于指定数组的维度，传入元组形式

### 5、级联
- np.concatenate((a1, a2, ...), axis=0, out=None)，沿着指定轴(axis = 0表示行，axis= 1表示列)连接对应的数列
 - a1,a2,...：表示待连接的数列；
 - axis：用于指定根据哪个轴进行连接，默认为0，即行；
- 注意：
 - 级联的参数必须是列表；
 - 维度必须相同；
 - 形状相符

````python
a1 = np.random.random(size = 3)
a2 = np.random.random(size = 3)
a3 = np.random.random(size = 3)
a4 = np.random.random(size = (2, 2))
a5 = np.random.random(size = (2, 2))
print(a1, a2, a3, a4, a5, sep = '\n==================================\n')
````

````python
con_a1 = np.concatenate((a1, a2, a3), axis = 0)
con_a2 = np.concatenate((a4, a5), axis = 0)
con_a3 = np.concatenate((a4, a5), axis = 1)
print(con_a1, con_a2, con_a3, sep = '\n=============================\n')
````

````python
# 生成猫的图片
cats1 = np.concatenate((cat, cat), axis = 0)
cats2 = np.concatenate((cat, cat), axis = 1)

plt.subplot(121)
plt.imshow(cats1)

plt.subplot(122)
plt.imshow(cats2)
plt.show()
````

#### 5.1、np.vstack与np.hstack
- np.vstack(tup)，垂直方向实现数列级联，即相当于np.concatenate中将axis设定为0；
- np.hstack(tup)，水平方向实现数列级联，即相当于np.concatenate中将axis设定为1；
- tup表示以元组形式将多个数组作为参数传入；

````python
# vertical垂直
np.vstack((a4, a5))
````

````python
# horizontal水平
np.hstack((a4, a5))
````

````python
# 将1行3列的数列转化为3行1列的数组
np.vstack(a1)
````

````python
# 将原2行2列的数组转化为一维数组
np.hstack(a4)
````

### 6、切分
- np.split(ary, indices_or_sections, axis = 0)，可实现对图片的切割
 - ary：将要被分割的数组；
 - indices_or_sections：若为一个整数n，则将数组分割为n等份；若为一维数组，则按照axis指定的轴进行分割；
 - axis：用于指定分割的轴
- np.vsplit(ary, indices_or_sections)：相当于axis = 0，按照行进行分割
- np.hsplit(ary, indices_or_sections)：相当于axis = 1，按照列进行分割

````python
ar1 = np.random.randint(1, 150, size = (5, 7))
ar2 = np.arange(9)
ar1, ar2
````

````python
# indices_or_sections为3，将原数组等分为三份
np.split(ar2, 3, axis = 0)
````

````python
# indices_or_sections为(1,3)，按行分割，分别[0,1)行,[1,3)行，[3,4)行，等同于np.vsplit(ar1, (1,3))
np.split(ar1, (1, 3), axis = 0) 
````

````python
# indices_or_sections为(1,3)，按列分割，等同于np.hsplit(ar1, (1,3))
np.split(ar1, (1, 3), axis = 1)
````

````python
cat2 = np.split(cat, (200, 400), axis = 0)[1]
cat3 = np.split(cat, (200, 400), axis = 1)[1]

plt.subplot(121)
plt.imshow(cat2)

plt.subplot(122)
plt.imshow(cat3)
plt.show()
````

### 7、副本
- 所有的赋值运算并不会为ndarray的任何元素创建副本，对赋值后对象的操作也对原来的对象有效
 - 如下例，通过令nd2 = nd1，并修改nd2中元素的值后，nd2和nd1在计算机的存储的位置并未发生变化(通过id方法观察)，通过修改nd2中的值会影响原数组的值；
 
- 为不改变原数组中的值，可通过**copy()**方法创建副本;
 - 如下例，通过使用copy()方法创建nd1的副本nd3，修改nd3中的值，并通过id方法查看二者在计算机中存储的位置可看出，nd1与nd3并不相同

````python
lst = [1, 2, 3, 4]
nd1 = np.array(lst)
nd1
````

````python
nd2 = nd1
nd2[2] = 111
nd1, nd2, id(nd1) == id(nd2)
````

````python
nd3 = nd1.copy()
nd3[2] = 111
nd1, nd3, id(nd3) == id(nd1)
````

## 六、广播机制
- 规则：
 - 矩阵计算时，若参与计算的矩阵维度不同，则在计算自动补齐为相同的维度进行计算；
 - 假定缺失的元素已有值进行填充进行计算；

## 七、ndarray的排序

### 7.1、冒泡排序

````python
nd = np.arange(10001)
np.random.shuffle(nd)
nd.shape, nd
````

````python
# 该方法使用两层for循环，时间复杂度和空间复杂度相对较高
def sortn1(a):
    for i in range(a.size):
        for j in range(i, a.size):
            if a[i] > a[j]:
                a[i], a[j] = a[j], a[i]
    return a

start = time.time()
sortn(nd)
end = time.time()
print('Processon time：%d' % (end - start) + 's')
````

````python
# 降低了时间和空间复杂度，效率更高
def sortn2(a):
    for i in range(a.size):
        min_index = np.argmin(a[i:]) + i
        a[i], a[min_index] = a[min_index], a[i]
    return a

start = time.time()
sortn2(nd)
end = time.time()
print('Processon time：%d' % (end - start) + 's')
````

### 7.２、快速排序
- np.sort()和ndarray.sort()的区别：
 - np.sort()：不该变输入
 - ndarray.sort()：本地处理，不占用空间，但改变输入，axis可以设置按照行或列进行排序

````python
n1 = np.array([2, 1, 6, 5, 4])
n1
````

````python
# 改变了原数组
n1.sort()
n1
````

````python
#不改变原数组
np.sort(n1), n1
````

### 7.3、部分排序
- np.partition(a, kth, axis = -1)，不查看所有数据的排序，而只是查看最大或最小的一部分数据
 - 当k为正数时，即得最小的k个数；
 - 当k为负数时，即得最大的k个数；

````python
n2 = np.random.randint(1, 15, size = 10)
n2
````

````python
np.partition(n2, 2), np.partition(n2, -2)
````
